## 2.2: Formatting the columns used to merge

The link column within the game_ratings data set must be cleaned to contain only the numerical variable to be able to join the 2 data sets together using this.

```python
# Define a function to clean the link column
def clean_link(link):
    return link.replace('/', '').replace('app', '')

# Apply the function to the 'link' column
game_ratings['link'] = game_ratings['link'].apply(clean_link)

# Save the cleaned dataset
game_ratings.to_csv('cleaned_game_ratings.csv', index=False)

print("Cleaned dataset:")
game_ratings
```

Ensure both the app_id and link columns are string variables to be able to join the table using this variable. The columns must be the same data to type to be able to join the tables. 

```python
# Convert 'app_id' to string in steam_games
steam_games['app_id'] = steam_games['app_id'].astype(str)
```

```python
#Ensure 'link' is of string type in game_ratings
game_ratings['link'] = game_ratings['link'].astype(str)
```

## 2.3: Merging the data sets

```python
# Merge the datasets on 'app_id' from steam_games and 'link' from game_ratings
steam_data = pd.merge(steam_games, game_ratings, left_on='app_id', right_on='link', how='inner')

# Display the first few rows of the merged dataset
print("Merged Dataset:")
steam_data
```

## 2.4: Cleaning the merged data set

Dropping all duplicated/uneccessary columns to save on storage space. 

```python
# Drop specific columns by name
columns_to_drop_3 = ['publisher_y', 'developer_y', 'Unnamed: 0', 'game', 'link', 'release', 'genres', 'categories', 'Unnamed: 0', 'peak_players', 'store_asset_mod_time', 'all_time_peak_date', 'all_time_peak', 'overall_review_count', 'about_description', 'overall_review_%', 'positive_reviews', 'negative_reviews']

# Drop the columns from the data frame by modifying the data frame in place without needing to reassign it
steam_data.drop(columns=columns_to_drop_3, inplace=True)

# Display the first few rows of the updated dataset
print("Merged Dataset:")
steam_data
```

```python
#Renaming the columns to improve usability
steam_data = steam_data.rename(columns={
    'overall_review': 'rating_desc',
    'developer_x': 'developer',
    'publisher_x': 'publisher',
})

# List all columns in the dataset
print("Columns in the dataset:")
print(steam_data.columns)
```

Renaming columns and changing column order to improve the usability and readability

```python
#Renaming the columns
steam_data = steam_data.rename(columns={
    'overall_review': 'rating_desc',
    'developer_x': 'developer',
    'publisher_x': 'publisher',
})
```

```python
# Reordering columns
steam_data = steam_data[['app_id', 'title', 'release_date', 'developer', 'publisher',
       'discounted_price', 'dlc_available', 'age_rating', 'win_support',
       'mac_support', 'linux_support', 'awards',  'players_right_now', '24_hour_peak',
        'primary_genre', 'store_genres','total_reviews', 'rating_desc','rating'
       ]]
```

Cleaning the discount_price columns to convert any 'Free' to 0.00, remove the ₹ and make it a numerical column. Whislt also converting from INR to GBP using the conversion rate of 1 INR = 0.0092926865 GBP as found on XE (2024).

```python
# Remove ₹ symbol from the 'price' column
steam_data['discounted_price'] = steam_data['discounted_price'].str.replace('₹', '')
```

```python
# convert any free values to 0.00
steam_data['discounted_price'] = steam_data['discounted_price'].replace('Free', 0.00)

# Summary statistics of numerical columns
print(steam_data.describe())
```

```python
# Convert the column to numeric, forcing any remaining non-numeric values to NaN
steam_data['discounted_price'] = pd.to_numeric(steam_data['discounted_price'], errors='coerce').fillna(0.00)

# Summary statistics of numerical columns
print(steam_data.describe())

# Filter rows where discounted_price is not equal to 0.0 or 'Free'
filtered_data = steam_data[(steam_data['discounted_price'] != 0.0) & (steam_data['discounted_price'] != 'Free')]

filtered_data
```

```python
#Converting from INR to GBP
conversion_rate = 0.0092926865

steam_data['discounted_price'] = steam_data['discounted_price'] * conversion_rate
```

Converting the release_date from DD MMM, YYYY format to dd/mm/yy. Some rows only contain the month and year of the release date and will be removed. 

```python
# Convert the 'release_date' column to datetime, handling mixed formats
steam_data['release_date'] = pd.to_datetime(steam_data['release_date'], errors='coerce')

# Format the dates to "dd/mm/yyyy"
steam_data['release_date'] = steam_data['release_date'].dt.strftime('%d/%m/%Y')
```

```python
# Checking for missing values in steam_data

missing_values = steam_data.isnull().sum()

print("Missing values for each column:\n", missing_values)
```

```python
# Dropping the 54 rows with missing values in the release_date column due to not containing the full dd mmm, yyyy format. 
steam_data.dropna(inplace = True)

# Display the first few rows of the updated dataset
print("\nUpdated DataFrame with dates in DD/MM/YYYY format:")
print(steam_data.head())
```

Removing the numerical values in parentheses from the primary_genre and store_genres columns

```python
# Remove the numerical values in parentheses from the 'primary_genre' column
steam_data['primary_genre'] = steam_data['primary_genre'].str.replace(r'\s*\(\d+\)', '', regex=True)

# Remove the numerical values in parentheses from the 'store_genres' column
steam_data['store_genres'] = steam_data['store_genres'].str.replace(r'\s*\(\d+\)', '', regex=True)
```

Checking the data type of all columns

```python
# Check the format (data type) of columns
print("\nData types of columns:")
print(steam_data.dtypes)
```

Formating all numerical columns to be to 2dp or integers.

```python
# Round specific float columns to 2 decimal places
columns_to_round = ['discounted_price', 'rating']
steam_data[columns_to_round] = steam_data[columns_to_round].round(2)
```

```python
# Convert object columns to numeric
numeric_columns = steam_data.select_dtypes(include=['object']).columns
steam_data['players_right_now'] = steam_data['players_right_now'].apply(pd.to_numeric, errors='coerce')
steam_data['24_hour_peak'] = steam_data['24_hour_peak'].apply(pd.to_numeric, errors='coerce')

# Convert numeric columns to integers
steam_data['players_right_now'] = steam_data['players_right_now'].astype('Int64')
steam_data['24_hour_peak'] = steam_data['24_hour_peak'].astype('Int64')
```

```python
# Check the format (data type) of columns
print("\nData types of columns:")
print(steam_data.dtypes)
```

Check fo any missing values due to data type conversions

```python
# Checking for missing values in steam_data

missing_values = steam_data.isnull().sum()

print("Missing values for each column:\n", missing_values)
```

```python
# Dropping the rows with missing values in the players_right_now and 24_hour_peak columns.
steam_data.dropna(inplace = True)

steam_data
```

# Section 3: Data Exploration

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Display basic information about the dataset
print(steam_data.info())

# Summary statistics of numerical columns
print(steam_data.describe())
```

```python
# Distribution of User Scores
plt.figure(figsize=(10, 6))
sns.histplot(steam_data['rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Relationship between Price and Average Playtime
plt.figure(figsize=(10, 6))
sns.scatterplot(x='discounted_price', y='24_hour_peak', data=steam_data, alpha=0.5)
plt.title('Relationship between Price and 24 Hour Peak')
plt.xlabel('Price')
plt.ylabel('24 Hour peak')
plt.show()
```

```python
# Example 3: Correlation Heatmap of Numerical Features
numerical_features = ['discounted_price', 'dlc_available', 'age_rating', 'awards', 'players_right_now', '24_hour_peak', 'total_reviews', 'rating']
plt.figure(figsize=(10, 8))
sns.heatmap(steam_data[numerical_features].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
#Boxplot of Ratings by Primary Genre
plt.figure(figsize=(12, 8))
sns.boxplot(x='primary_genre', y='rating', data=steam_data)
plt.title('Ratings Distribution by Primary Genre')
plt.xticks(rotation=45)
plt.xlabel('Primary Genre')
plt.ylabel('Rating')
plt.show()
```
