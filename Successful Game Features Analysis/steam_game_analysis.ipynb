# Section 4: Random Forest Algorithm

## Section 4.1: Preprocessing

As the sklearn algortithm that will be used in Python for the random forest algorithm can't handle non-numeric data one hot encoding & frequency encoding must be used to convert categorical and non-numeric data.


```python
# Importing relevant modules
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score

# Count the frequency of each value in each object type column

# Filter for object (categorical) columns
object_columns = steam_data.select_dtypes(include=['object'])

# Count unique values for each categorical column
unique_counts = object_columns.nunique()

print(unique_counts)
```

```python
# Frequency Encoding of developer, publisher and store genres due to large number of unique variables
# Frequency Encoding Function
def frequency_encoding(steam_data, column):
    freq = steam_data[column].value_counts()
    steam_data[column] = steam_data[column].map(freq)
# Apply frequency encoding
frequency_encoding(steam_data, 'developer')
frequency_encoding(steam_data, 'publisher')
frequency_encoding(steam_data, 'store_genres')
```

```python
# One-hot encoding for 'primary_genre'
encoded_data = pd.get_dummies(steam_data, columns=['primary_genre'])

# Display the resulting DataFrame
print(encoded_data.head())
```

## Section 4.2: Creating the target variable and test and train sets

As the success metric has a rating of >= 85 a new target variable column must be created for successful games (rating >= 85) to be marked as 1 and unsuccessful games (rating <85) to be marked as 0. 

```python
# Creating the success metric column
encoded_data['success_metric'] = encoded_data['rating'] >= 85

# Convert the boolean success_metric to integer (1 for True, 0 for False)
encoded_data['success_metric'] = encoded_data['success_metric'].astype(int)
```

```python
# Define the target variable
y = encoded_data['success_metric']  # Binary target: 1 for success, 0 for not successful

# Define the features (exclude the target variable)
X = encoded_data.drop(['success_metric','rating'], axis=1)  
# 'rating' is also dropped to avoid data leakage

# Check X Data
X
```

'y' is now a binary variable with 1 representing a rating of 85 or over and 0 representing a rating below 85
'x' now contains all other features within the data set and will be used to predict whether a game will be successful. 

To evaluate the model's performance, the data set must be split into two sets: a test and a training data set. The data sets are also scaled to ensure each feature has a uniform contribution to the model and reduce the chance of bias in the splits.

```python
# Split the data into training and testing sets (30% of the data will be used for the test data set)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Identify numerical columns
numerical_cols = X_train.select_dtypes(include=['int64', 'float64']).columns

# Scale only the numerical features
scaler = StandardScaler()
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])

X_train
```

## Section 4.3: Training the Random Forest Model & Feature Tuning

```python
from sklearn.model_selection import GridSearchCV

# Initialise the model for feature tuning
rf_model = RandomForestClassifier(random_state=42)


# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees
    'max_depth': [10, 20, 30, 40],    # Depth of the trees
    'min_samples_split': [2, 5, 10],   # Minimum number of samples required to split an internal node
    'class_weight': [None, 'balanced', 'balanced_subsample']  # Handling class imbalance
}

# Perform Grid Search
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='f1_macro', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Best model and its parameters
best_model = grid_search.best_estimator_
print("Best parameters:", grid_search.best_params_)

```

```python
# Initialise the model
rf_model = RandomForestClassifier(min_samples_split=10,  # Set this to your desired value
                                  n_estimators=300,      # Number of trees
                                   max_depth=20,        # Depth of the trees
                                  class_weight='balanced',  # Address class imbalance
                                  random_state=42)       # Seed for reproducibility)

# Train the model
rf_model.fit(X_train, y_train)
```

## Section 4.4: Evaluating The Model

### Section 4.4.1: Accuracy, Precision, Recall and F1 Scores

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Calculate precision
precision = precision_score(y_test, y_pred)

# Calculate recall
recall = recall_score(y_test, y_pred)

# Calculate F1 score
f1 = f1_score(y_test, y_pred)

# Print the results
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=rf_model.classes_)

# Plot the confusion matrix
plt.figure(figsize=(10, 7))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rf_model.classes_)
disp.plot(cmap=plt.cm.Blues, values_format='d')  # Display in a blue colormap with integer format
plt.title('Confusion Matrix')
plt.show()
```

### Section 4.4.2: Cross-Validation of Accuracy, Precision, Recall and F1 Scores

Cross-validation will be used to check the consistency of the model’s performance across different subsets of the data. This helps in assessing whether the performance is consistent or just due to a specific train-test split.

```python
import numpy as np
from sklearn.model_selection import cross_validate
from sklearn.datasets import load_iris
# Define the metrics
scoring = ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']

# Perform cross-validation
results = cross_validate(rf_model, X, y, cv=5, scoring=scoring, return_train_score=False)

# Print results
print("Accuracy scores: ", results['test_accuracy'])
print("Precision scores: ", results['test_precision_macro'])
print("Recall scores: ", results['test_recall_macro'])
print("F1 scores: ", results['test_f1_macro'])

# Calculate mean and standard deviation for each metric
print("\nMean Accuracy: {:.2f} ± {:.2f}".format(np.mean(results['test_accuracy']), np.std(results['test_accuracy'])))
print("Mean Precision: {:.2f} ± {:.2f}".format(np.mean(results['test_precision_macro']), np.std(results['test_precision_macro'])))
print("Mean Recall: {:.2f} ± {:.2f}".format(np.mean(results['test_recall_macro']), np.std(results['test_recall_macro'])))
print("Mean F1: {:.2f} ± {:.2f}".format(np.mean(results['test_f1_macro']), np.std(results['test_f1_macro'])))
```

### Section 4.4.3: Learning Curves of Accuracy, Precision, Recall and F1 Scores

```python
import numpy as np
from sklearn.model_selection import learning_curve
from sklearn.metrics import make_scorer, precision_score, recall_score, f1_score

# Define custom scoring functions
precision_scorer = make_scorer(precision_score, average='macro')
recall_scorer = make_scorer(recall_score, average='macro')
f1_scorer = make_scorer(f1_score, average='macro')

# Function to plot learning curves
def plot_learning_curve(title, estimator, X, y, scoring, cv=5):
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, scoring=scoring, n_jobs=-1,
        train_sizes=np.linspace(0.1, 1.0, 10)
    )
    
    # Calculate mean and std for train and test scores
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)
    
    # Plot learning curve
    plt.figure(figsize=(10, 7))
    plt.plot(train_sizes, train_mean, 'o-', color='r', label='Training score')
    plt.plot(train_sizes, test_mean, 'o-', color='g', label='Validation score')
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='r')
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color='g')
    plt.title(title)
    plt.xlabel('Training examples')
    plt.ylabel(scoring)
    plt.legend(loc='best')
    plt.grid()
    plt.show()

# Plot learning curves for Accuracy, Precision, Recall, and F1 Score
plot_learning_curve('Learning Curve (Accuracy)', rf_model, X, y, scoring='accuracy')
plot_learning_curve('Learning Curve (Precision)', rf_model, X, y, scoring='precision')
plot_learning_curve('Learning Curve (Recall)', rf_model, X, y, scoring='recall')
plot_learning_curve('Learning Curve (F1 Score)', rf_model, X, y, scoring='f1')
```

### Section 4.4.4: ROC Curve

```python
# Create ROC curve

# Predict probabilities for the positive class (1 for success)
y_pred_proba = rf_model.predict_proba(X_test)[:, 1]

from sklearn.metrics import roc_curve, roc_auc_score

# Calculate the false positive rate, true positive rate, and thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Calculate the AUC score
auc_score = roc_auc_score(y_test, y_pred_proba)

print(f'AUC Score: {auc_score}')


```

```python
# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')  # Diagonal line for reference
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
